```python
#Q2.

# A scatter plot is a type of data visualization that uses dots or markers to represent individual data points on a two-dimensional plane.

import matplotlib.pyplot as pt
```

```python
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
x
```

```python
y
```

```python
pt.scatter(x,y)
pt.xlabel("X Values")
pt.ylabel("Y Values")
pt.title("Scatter Plot of Gaussian Distribution around a mean of 3")
```

```python
#Q3.
# 'subplot()' is used to create multiple plots within a single figure.
# It allows one to create a grid of rows and columns of subplots and specify which subplot one wants to work on based on its position in the grid.

x = np.array([0,1,2,3,4,5])

#Line 1
y1 = np.array([0,100,200,300,400,500])
pt.subplot(2,2,1)
pt.plot(x,y1)
pt.title("Line 1")

#Line 2
y2 = np.array([50,20,40,20,60,70])
pt.subplot(2,2,2)
pt.plot(x,y2)
pt.title("Line 2")

#Line 3
y3 = np.array([10,20,30,40,50,60])
pt.subplot(2,2,3)
pt.plot(x,y3)
pt.title("Line 3")

#Line 4
y4 = np.array([200,350,250,550,450,150])
pt.subplot(2,2,4)
pt.plot(x,y4)
pt.title("Line 4")

pt.tight_layout()

pt.show()
```

```python
#Q4.

# A bar plot is a graphical representation using bars to compare data across categories. 

#It's used to easily compare, display frequency, and show trends in categorical data.

company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
pt.bar(company,profit)
pt.xlabel("Organization Name")
pt.ylabel("Profit Gained")
pt.title("Bar Graph representing profit margin gained by tech giants")
pt.grid()
```

```python
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
pt.barh(company,profit)
pt.xlabel("Profit Gained")
pt.ylabel("Organization Name")
pt.title("Bar Graph representing profit margin gained by tech giants")
pt.grid()
```

```python
#Q5.

# A box plot is a graphical representation that displays the distribution of a dataset's values. It shows the median, quartiles, and potential outliers in a concise manner.

# They are used to identify the spread, skewness, and presence of outliers in data, making them valuable for data comparison.

box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
```

```python
data = [box1,box2]
pt.boxplot(data)
pt.show()
```
